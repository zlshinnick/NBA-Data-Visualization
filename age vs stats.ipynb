```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("./all_seasons.csv")
```

```python
df = df.groupby("age").mean()

df=df[["pts","reb","ast"]]

plt.figure(figsize=(16,8))
plt.xticks(rotation=45)
sns.lineplot(data=df)
font1 = {'family':'arial','color':'black','size':20}
font2 = {'family':'arial','color':'black','size':12}

plt.xlabel("Age of Player", fontdict = font2)
plt.ylabel("Average Statistics", fontdict = font2)
plt.title("Age vs Players Performance",fontdict=font1)
```
