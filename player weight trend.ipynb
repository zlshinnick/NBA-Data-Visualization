```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("./all_seasons.csv")
```

```python
df = df.groupby("season").mean()

df=df[["player_weight"]]
```

```python
plt.figure(figsize=(20,10))

sns.lineplot(df.index,df.player_weight)
plt.xticks(rotation=-30)

font1 = {'family':'arial','color':'black','size':20}
font2 = {'family':'arial','color':'black','size':12}

plt.xlabel("Season", fontdict = font2)
plt.ylabel("Weight(KG)", fontdict = font2)
plt.title("Average Player Weight per Season",fontdict = font1)
```

```python
plt.figure(figsize=(20,10))

sns.lineplot(df.index,df.player_height)
plt.xticks(rotation=-30)

font1 = {'family':'arial','color':'black','size':20}
font2 = {'family':'arial','color':'black','size':12}

plt.xlabel("Season", fontdict = font2)
plt.ylabel("Height(CM)", fontdict = font2)
plt.title("Average Player Height per Season",fontdict = font1)
```
