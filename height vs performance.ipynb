```python
 filter_out = players.pts!=0
players = players[filter_out]
```

```python
plt.figure(figsize=(15,10))
ax = sns.boxplot(x=players.height_range,y=players.pts)
ax.set_title('Points by Height Range')
ax.set_ylabel ('Points Per Game')
ax.set_xlabel ('Height Range')
```

```python
plt.figure(figsize=(15,10))
ax =sns.boxplot(x=players.height_range,y=players.ast)
ax.set_title('Assists by Height Range')
ax.set_ylabel ('Assists Per Game')
ax.set_xlabel ('Height Range')
```

```python
plt.figure(figsize=(15,8))
ax= sns.boxplot(x=players.height_range,y=players.reb)
ax.set_title('Rebounds by Height Range')
ax.set_ylabel ('Rebounds Per Game')
ax.set_xlabel ('Height Range')




```

```python
plt.figure(figsize=(20,10))

ax= sns.lineplot(x=players.height_range,y=players.net_rating).set(title='Net Rating by Height Range')

font1 = {'family':'arial','color':'black','size':22}
font2 = {'family':'arial','color':'black','size':12}

plt.xlabel("Height of Players", fontdict = font2)
plt.ylabel("Average Net Rating", fontdict = font2)
plt.title("Height Vs Net Rating",fontdict = font1)
```
