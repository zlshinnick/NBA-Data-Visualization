```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("./all_seasons.csv")
```

```python
def visualize_average_trend(col):
    players = pd.DataFrame(df.groupby(['team_abbreviation', 'season'])[col].mean())
    players = players.rename({'player_name': f'Average {col}'}, axis=1)
    players.reset_index().head()
    plt.figure(figsize=(20,10))

    sns.lineplot('season', col, data=players)
    
    title = (' '.join(col.split('_'))).title()

    plt.xticks(rotation=-30)
    plt.ylabel(title)
    plt.title(f'Average {title} per Season', fontsize=20)
 
    plt.show()
```
