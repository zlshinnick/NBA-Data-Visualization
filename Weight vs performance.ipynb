```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("./all_seasons.csv")
```

```python
filter_out = df.player_weight<155
df = df[filter_out]
```

```python
df
```

```python


df['weight_range'] = 5*round(df.player_weight/5)



# Group by draft year
players_df = df.groupby("weight_range").mean()
players_df = df.groupby("weight_range").mean()

# Selecting only age column
players_df = players_df[["pts","ast","reb"]]
```

```python
df
```

```python
plt.figure(figsize=(16,8))
plt.xticks(rotation=45)
sns.lineplot(data=players_df)
font1 = {'family':'arial','color':'black','size':20}
font2 = {'family':'arial','color':'black','size':12}

plt.xlabel("Weight of Player (KG)", fontdict = font2)
plt.ylabel("Average of Stats", fontdict = font2)
plt.title("Weight vs Performance data", fontdict = font1)
```
