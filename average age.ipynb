```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("./all_seasons.csv")
```

```python
df = df.groupby("season").mean()

df=df[["age"]]

plt.figure(figsize=(16,8))
plt.xticks(rotation=45)
sns.lineplot(data=df)
font1 = {'family':'arial','color':'black','size':20}
font2 = {'family':'arial','color':'black','size':12}

plt.xlabel("Season", fontdict = font2)
plt.ylabel("Average Age", fontdict = font2)
plt.title("Average Age of NBA Players from 1996-220",fontdict = font1)
```
